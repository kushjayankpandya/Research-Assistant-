```python
def download_image(img_url,videoid,search):
    response = requests.get(img_url)
    if response.status_code:
        fp = open(f'{search}/{videoid}.png', 'wb')
        fp.write(response.content)
        fp.close()

def ocr(videoid,search):
    img = Image.open(f'{search}/{videoid}.png',)
    text1 = pytesseract.image_to_string(img,lang='eng',config='--psm 6 --oem 3 tessedit_char_whitelist #abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789')
    text2 = pytesseract.image_to_string(img,lang='eng',config='--psm 12 --oem 3 tessedit_char_whitelist #abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789')
    text3 = pytesseract.image_to_string(img,lang='eng',config='--psm 4 --oem 3 tessedit_char_whitelist #abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789')
    text = text1 + text2 + text3
    pattern = r'(\#\w+)'
    output = set(re.findall(pattern, text))
    return output
```

```python


def page_extractor(items,search):
    
    dataset = []
    for videos in items:
        tupple = {}
        
        tupple['videoid'] = videos['id']['videoId']
        print('video id - ',tupple['videoid'])
        tupple['publishedAt'] = videos['snippet']['publishedAt']
        tupple['thumbnail_url'] = videos['snippet']['thumbnails']['high']['url']
        download_image(tupple['thumbnail_url'],tupple['videoid'],search)
        tupple['hashtags'] = ocr(tupple['videoid'],search)
        if len(tupple['hashtags']) != 0:
            dataset.append(tupple)
            print(tupple['hashtags'])

    return dataset 
```

```python
from msilib.schema import Error


def youtube_data_extractor(search_keyword,pages):
    
    dataset = []
    token = ''
    hashcounter = 0
    counter = 0
    
    for i in range(pages):
    
        try:
            
            lisearch_keyword = youtube_object.search().list(q = search_keyword, part = "id, snippet",
                                                    maxResults = 50,pageToken = token,type='video').execute()
            
            
            token = lisearch_keyword['nextPageToken']
            
            counter += 1
            print(f'page no:- {counter}')

            search_keyword_word = search_keyword.replace(' ','_')
            os.system(f'mkdir {search_keyword_word}')
            results = page_extractor(lisearch_keyword['items'],search_keyword_word)

            results = {tupple['videoid']:tupple  for tupple in results}

            if len(results) == 0:
                continue
            else:
                hashcounter+=len(results)

            statistics = youtube_object.videos().list(id=list(results.keys()),part="statistics").execute()

            for stats in statistics['items']:
                videoid = stats['id']
                tupple = results[videoid]
                tupple['viewCount'] = stats['statistics']['viewCount']
                tupple['commentCount'] = stats['statistics']['commentCount']
                dataset.append(tupple)
        except:
            return dataset

    print(f'catch rate - {hashcounter/(pages*50)} %')


    df = pd.DataFrame.from_dict(dataset)
    df.to_csv(search_keyword+'.csv')

    
values = youtube_data_extractor('arnab goswami debates republic tv',20)        
```
